```python
# Se importa las librerias instaladas
from GoogleNews import GoogleNews
from newspaper import Article
from newspaper import Config
# Sera util en el momento de usar Article
import nltk
nltk.download('punkt')
# Pandas para la manipulacion de datos
import pandas as pd
# Librerias para visualización
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Se instancia el objeto de GoogleNews
googlenews = GoogleNews()
# Se personaliza el idioma a español
googlenews.set_lang('es')
# Se personaliza el tiempo de busqueda  (mm/dd/yyyy)
googlenews.set_time_range('01/01/2020','12/31/2023')
# Tipo de codificación
googlenews.set_encode('utf-8')
# Se realiza una busqueda personalizada
topico = 'colombia'
googlenews.search(topico)
```

Hay un límite de búsqueda en la cantidad de noticias en esta API, debido a bloqueos de IP. Sin embargo se tiene certeza que se puede evitar este inconveniente automaticamente. Por el momento para la creación del dataset manualmente se ha desconectado y borrado el tiempo de ejecición de la maquina virtual de google colab para obtener nuevas IP's y realizar multiples busquedas de diferentes topicos: 'inteligencia artificial', 'python colombia'

```python
# Se intentará extraer 200 páginas con el tema de interés
for i in range(2, 200):
    # Obtener la  página de resultados de Google News
    googlenews.getpage(i)

    # Obtener los resultados de la  página actual
    result = googlenews.result()

    # Crear un DataFrame a partir de los resultados
    df = pd.DataFrame(result)
```

```python
# Se eliminan duplicados para evitar noticias repetidas
df.drop_duplicates()
```

Siguiendo el procedimiento mencionado anteriormente y empleando Google News, no se extrae directamente el texto de la noticia, sino el enlace donde está alojada. Para la fase siguiente de descargar el contenido de la noticia, se utiliza newspaper3k

Se define un agente de usuario (User Agent) simulado que se utilizará al realizar solicitudes web. Esto puede ser útil para evitar problemas de acceso a ciertos sitios web.

```python
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'
config = Config()
config.browser_user_agent = user_agent
```

```python
# lista para almacenar diccionarios de información de los artículos
list = []

# Iterar a través de las filas del DataFrame df
for ind in df.index:
    try:
        # Inicializar un diccionario para almacenar información del artículo actual
        dict = {}

        # Crear un objeto Article con la URL del artículo y la configuración personalizada
        article = Article(df['link'][ind], config=config)

        # Descargar el contenido del artículo desde la URL
        article.download()

        # Analizar el contenido del artículo para extraer información
        article.parse()

        # Realizar un análisis de procesamiento de lenguaje natural (NLP) en el contenido
        article.nlp()

        # Agregar información del artículo al diccionario
        dict['Date'] = df['date'][ind]
        dict['Media'] = df['media'][ind]
        dict['Title'] = article.title
        dict['Article'] = article.text
        dict['Summary'] = article.summary

        # Agregar el diccionario a la lista
        list.append(dict)
    except Exception as e:
        # Manejo de errores
        # La pagina no existe y continuar con la siguiente URL
        continue

# Crear un DataFrame a partir de la lista de diccionarios
news_df = pd.DataFrame(list)
# Filtrar donde no fue posible extraer la información
news_df = news_df[news_df['Article'].apply(lambda x: bool(x))]
# Mostrar el DataFrame resultante
display(news_df)
```

```python
# Creación de la ruta para guardar el dataset de noticias
path = f"/content/{topico}.csv"
# Se guarda el dataset
news_df.to_csv(path)
```
